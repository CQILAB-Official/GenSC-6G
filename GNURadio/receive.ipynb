```python
# Load .bin file and show head
import numpy as np

data = np.fromfile('received_data.bin', dtype=np.float32)

np.set_printoptions(threshold=np.inf, linewidth=np.inf)
# Limit the output to the first 100,000 elements
limited_data = data[:100000]

# Save the first 100,000 elements of the data to a .txt file in the specified format
with open('received_data_limited.txt', 'w') as file:
    file.write(f"array({limited_data})\n")

```

```python
# Load .bin file and show head
import numpy as np

data = np.fromfile('transmit_data.bin', dtype=np.float32)

np.set_printoptions(threshold=np.inf, linewidth=np.inf)
# Limit the output to the first 100,000 elements
limited_data = data[:100000]

# Save the first 100,000 elements of the data to a .txt file in the specified format
with open('transmit_data_limited.txt', 'w') as file:
    file.write(f"array({limited_data})\n")

```

```python
import matplotlib.pyplot as plt
from scipy.signal import welch

freqs, psd = welch(data, fs=2e6, nperseg=1024)

# Plot the PSD
plt.figure(figsize=(10, 6))
plt.semilogy(freqs, psd)
plt.title('Power Spectral Density of Captured Noise')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.show()
```
